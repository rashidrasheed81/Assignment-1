## np.random


```python
# np.random.randint

# toss a coin
x = np.random.randint(2)
print(x)

# toss a coin 5 times
y = np.random.randint(2,size=5)
print(y)

# roll a dice
z = np.random.randint(1,7)
print(z)

# roll a dice 10 times
a = np.random.randint(1,7,size=10)
print(a)
```

```python
# toss a coin
a = np.random.choice(["Head","Tail"])
print(a)

# toss a coin 5 times
b = np.random.choice(["Head","Tail"],size=5)
print(b)

# roll a dice
c = np.random.choice([1,2,3,4,5,6])
print(c)

# roll a dice 10 times
d = np.random.choice([1,2,3,4,5,6],size=10)
print(d)
```

```python
# np.random.binomial

# number of successes (number of heads) in 10 coin flips
a = np.random.binomial(10,0.5)
print(a)

# number of successes (number of heads) in 10 coin flips - biased coin with a 0.8 probability of obtaining heads
b = np.random.binomial(10,0.8)
print(b)

# we can obtain approximated probabilities by simulating a huge number of flips probability of obtaining 4 head in 10 flips 
flips = np.random.binomial(10,0.5,size=int(1e6))
(flips==4).mean()
```

```python
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
print(s)

#Verify the mean and the variance:
a = abs(sigma - np.std(s, ddof=1)) < 0.01
b = abs(mu - np.mean(s)) < 0.01
print(a,b)
```

```python
# np.random.poisson and then graph using matplotlib

s = np.random.poisson(5, 10000)
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 14, density=True)
plt.show()
```

```python
# np.random.power()
a = 5. # shape
samples = 1000
s = np.random.power(a, samples)
count, bins, ignored = plt.hist(s, bins=30)
x = np.linspace(0, 1, 100)
y = a*x**(a-1.)
normed_y = samples*np.diff(bins)[0]*y
plt.plot(x, normed_y)
plt.show()
```

```python
# np.random.uniform()

s = np.random.uniform(-1,0,1000)
count, bins, ignored = plt.hist(s, 15, density=True)
plt.plot(bins, np.ones_like(bins), linewidth=2, color='r')
plt.show()
```

```python
#numpy.ndarray.byteswap()
a = np.array([1, 256, 8755], dtype = np.int16) 

print ('Our array is: ', a)   

print ('Representation of data in memory in hexadecimal form: ', map(hex,a))  

# byteswap() function swaps in place by passing True parameter 

print ('Applying byteswap() function: ',a.byteswap(True)) 

print ('In hexadecimal form: ',map(hex,a))
```

```python
#numpy.sort()
a = np.array([[3,7],[9,1]]) 

print ('Our array is: ',a,'\n') 

print ('Applying sort() function: ',np.sort(a),'\n') 
  
print ('Sort along axis 0:') 
print (np.sort(a, axis = 0)) 
print ('\n')  

# Order parameter in sort function 
dt = np.dtype([('name', 'S10'),('age', int)]) 
a = np.array([("raju",21),("anil",25),("ravi", 17), ("amar",27)], dtype = dt) 

print ('Our array is: ',a,'\n') 

print ('Order by name:') 
print (np.sort(a, order = 'name'))
```

```python
#numpy.argsort()
x = np.array([3, 1, 2]) 

print ('Our array is: ',x,'\n') 

print ('Applying argsort() to x: ') 
y = np.argsort(x) 
print (y,'\n') 

print ('Reconstruct original array in sorted order:',x[y],'\n' ) 

print ('Reconstruct the original array using loop:') 
for i in y: 
   print (x[i])
```

```python
#numpy.lexsort()
nm = ('raju','anil','ravi','amar') 
dv = ('f.y.', 's.y.', 's.y.', 'f.y.') 
ind = np.lexsort((dv,nm)) 

print ('Applying lexsort() function:',ind,'\n') 

print ('Use this index to get sorted data:') 
print ([nm[i] + ", " + dv[i] for i in ind]) 
```

```python
#numpy.nonzero()
a = np.array([[30,40,0],[0,20,10],[50,0,60]]) 

print ('Our array is:','\n') 

print ('Applying nonzero() function:') 
print (np.nonzero (a))
```

```python
#numpy.where()
x = np.arange(9.).reshape(3, 3) 

print ('Our array is: ',x) 

print ('Indices of elements > 3') 
y = np.where(x > 3) 
print (y)  

print ('Use these indices to get elements satisfying the condition') 
print (x[y])
```

```python
#numpy.extract()
x = np.arange(9.).reshape(3, 3) 

print ('Our array is: ',x)

# define a condition 
condition = np.mod(x,2) == 0 

print ('Element-wise value of condition',condition) 

print ('Extract elements using condition') 
print (np.extract(condition, x))
```
